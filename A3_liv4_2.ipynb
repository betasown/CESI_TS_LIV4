```python
from numpy import *

mess_ascii = input("veuillez entrer le message que vous voulez envoyer")
mess_bin = ''.join(format(ord(char), '08b') for char in mess_ascii)         # conversion Ascii (caractère) / Binaire (octet)
print(mess_ascii)
print(mess_bin)
```

```python
mess_bin_list = list(mess_bin)      # passe mess_bin d'un  à une liste
print(mess_bin_list)
```

```python
for i in range (len(mess_bin_list)):
    mess_bin_list [i] = int(mess_bin_list [i])  # passe d'une liste de caractere à une liste d'integer
    
print(mess_bin_list)

rev = mess_bin_list.reverse()                   # .reverse permet d'inverser toutes les valeurs d'une liste
print(mess_bin_list)
```

```python
G = 9
G_en_binaire = str(1001)                # def une valeur aleatoire pour le diviseur
rev = mess_bin_list.reverse()           # .reverse permet d'inverser toutes les valeurs d'une liste
print(mess_bin_list)
new = [0, 0, 0, 0,]
mess_bin_list_p = mess_bin_list+new     # rajoute new à la fin de la liste
print(mess_bin_list_p)
```

```python
new_data_bin = ""                       # initialisation
for elem in mess_bin_list_p:
    new_data_bin += str(elem)
print(new_data_bin)

def binaryToDecimal(n):                 # fonction qui converti une valeur binaire en une valeur numérique
    return int(n,2)
print(binaryToDecimal(new_data_bin))
print(binaryToDecimal(G_en_binaire))
```

```python
reste = binaryToDecimal(new_data_bin) % binaryToDecimal(G_en_binaire)   #reste = valeur numérique du message % diviseur
print(reste)
```

```python
reste_bin = format(reste, "b")      # passe le reste en binaire
print(reste_bin)
```

```python
reste_bin_list = list(reste_bin)                    # transforme en liste de caracteres
for i in range (len(reste_bin_list)):               # fonction qui transforme en liste d'integer
    reste_bin_list [i] = int(reste_bin_list [i])
print(reste_bin_list)
```

```python
mess_crc = mess_bin_list_p + reste_bin_list         # rajoute le reste à la liste
print(mess_crc)
```

```python
fanion=[0,1,1,1,1,1,1,0]                # fanion à def
adresse=[0,0,0,0,1,1,1,1]               # adresse à def
control=reste_bin_list
while len(control)<8:                   # fonction qui rajoute des 0 pour que control ait une taille de 8
    control = [0]+control
mess_a_envoyer = fanion + adresse + control + mess_crc + fanion # assemblage de la trame
print(mess_a_envoyer)
print(control)
```

```python
mess_code = []                  # def liste vide
current_state = 0               # État initial

for bit in mess_a_envoyer:      # fonction qui code en NRZI
    if bit == 0:
        mess_code.append(current_state)
    else:
        current_state = 0 if current_state == 1 else 1
        mess_code.append(current_state)

print(mess_code)                # Affiche la séquence codée en NRZI
```

```python
import matplotlib.pyplot as plt
import numpy as np
import sounddevice as sd
```

```python
M = mess_code                        # Message binaire I

Fe = 44100                           # Fréquence d'échantillonnage
Fp = 8000                            # Fréquence de l'onde porteuse 


baud = 300                           # Débit souhaité sur le canal de transmission exprimé en bit/s
Nbits = len(M)                       # Nombre de bits initial (taille du message I)
Ns = int(Fe/baud)                    # Nombre de symboles par bit (Fréq d'echan / Débit binaire) 

```

```python
N = int(Nbits*Ns)                    # Nombre de bits total à moduler (Nombre de symboles par bit * Nombre de bits)

# On génère le message binaire dupliqué
M_duplique = np.repeat(M, Ns)           # A COMPLETER On fonction repeat ou fonction tile de numpy

# On génère le vecteur temps t1
t1 = np.linspace(0.0, Ns/Fe, Ns)

# On génère le vecteur temps t
t = np.linspace(0.0, N/Fe, N)



# Génération des 2 porteuse P1 pour le bit 1 et P2 pour le bit 0
A1 = 1                   # A COMPLETER - Amplitude de la porteuse 1 du bit 1                           
A2 = 1                   # A COMPLETER - Amplitude de la porteuse 2 du bit 0
fp1 = 500                  # A COMPLETER - fréquence de la porteuse du bit 1
fp2 = 2000                  # A COMPLETER - fréquence de la porteuse porteuse du bit 0



# On génère des porteuses d'une longueur correspondant à Ns

P1 = A1*np.sin(2*np.pi*fp1*t1)                   # A COMPLETER - porteuse du bit  1  
P2 = A2*np.sin(2*np.pi*fp2*t1)                   # A COMPLETER - Porteuse du bits 0



# On réalise la modulation FSK
FSK = []       # Initialisation d'une liste nommée FSK

                        # A COMPLETER
for n in range(0, len(M)):
    if (M[n] == 1):
        FSK.extend(P1)
    else:
        FSK.extend(P2)
        
        

# Affichage du Message binaire, la porteuse et le résultat de la modulation ASK
                        # A COMPLETER


#Affichage du message M(t)

plt.figure (figsize = (10,6))
plt.plot(t,M_duplique,'black')
plt.title('Message M(t)')
plt.xlabel('temps (s)')
plt.ylabel('Amplitude')
plt.grid()

#Affichage de la porteuse P1(t)

plt.figure (figsize = (10,6))
plt.plot(t1,P1,'b')
plt.title('Porteuse P1(t)')
plt.xlabel('temps (s)')
plt.ylabel('Amplitude')
plt.grid()

#Affichage de la porteuse P2(t)

plt.figure (figsize = (10,6))
plt.plot(t1,P2,'b')
plt.title('Porteuse P2(t)')
plt.xlabel('temps (s)')
plt.ylabel('Amplitude')
plt.grid()

#Affichage du résultat de la modualtion AM(t)
      
plt.figure (figsize = (10,6))
plt.plot(t,FSK,'r')
plt.title('Modualtion FSK')
plt.xlabel('temps (s)')
plt.ylabel('Amplitude')
plt.grid()
```

```python
# A COMPLETER 

####Voir le schéma de principe de la démodulation FSK et suivre les étapes décrites dans le texte) ###

Porteuse1 = np.tile(P1, Nbits)      # np.tile permet de mettre sur une base de temps t, car la porteuse est sur t1
Porteuse2 = np.tile(P2, Nbits)

bit1 = FSK*Porteuse1
bit0 = FSK*Porteuse2

plt.figure (figsize = (10,6))
plt.plot(t,bit1)
plt.title('Bit1')
plt.xlabel('temps (s)')
plt.grid()

y1 = []
y2 = []

i=0

for i in range(0, N, Ns):
    y1.append(np.trapz(bit1[i:i+Ns]))
    y2.append(np.trapz(bit0[i:i+Ns]))        

    # Remarque sur la fonction trapz :
    
    # On se contente d'intégrer sur un seul axe et c'est suffisant ici
    # si les données ne sont pas réparties uniformément par l'échantillonnage, il faut aussi intégrer sur l'axe des abscisses
    # np.trapz(Produit[i:i+Ns], t[i:i+Ns])
```

```python

# Si abs(y1) > abs(y2) alors on a reçu un 1 (True), sinon un 0 (False) (y1 et y2 résultats obtenus dans l'étape précédente)

# A COMPLETER 

message_demodule_FSK =  []

for ii in range(0, len(y1)):
    if abs(y1[ii]) > abs(y2[ii]):
        message_demodule_FSK.extend([int(1)])
    if abs(y1[ii]) <= abs(y2[ii]):
        message_demodule_FSK.extend([int(0)])
```

```python
print("Le message binaire envoyé est : \n ", M)
print("Le message binaire démodulé est : \n ", message_demodule_FSK)

# Calculer si erreur de réception

# La lsite Erreur contient True si la transmission s'est bien passée, False sinon

Erreur = []               # Initialiser une liste
Erreur = (np.array(M) == np.array(message_demodule_FSK))                # A COMPLETER 

print ("Erreur de réception :\n", Erreur)
        
    
```

```python
mess_code = message_demodule_FSK

decoded_message = []
current_state = 0           # État initial

for state in mess_code:     # decode le NRZI
    if state == current_state:
        decoded_message.append(0)
    else:
        decoded_message.append(1)
        current_state = 1 if current_state == 0 else 0

print(decoded_message)      # Affiche le message décodé
```

```python
mess_code = decoded_message

# Enlever les deux premiers octets et le dernier octet
mess_code_sans_bordures = mess_code[16:-8]

print(mess_code_sans_bordures)
```

```python
mess_code_control = mess_code_sans_bordures

control = mess_code_control[:8]  # Prend les 8 premiers bits et les stocke dans control
del mess_code_control[:8]        # Supprime les 8 premiers bits de la liste d'origine

print("Control:", control)
print("Mess_code sans la partie control:", mess_code_control)
```
